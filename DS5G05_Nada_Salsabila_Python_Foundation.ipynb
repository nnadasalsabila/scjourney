#### 5. Buatlah status BMI untuk setiap siswa dari kategori berikut, lalu simpan data statusnya dengan nama kolom Status BMI.

Kategori:
*   Normal: 18.5-24.9
*   Overweight: 25.0-29.9
*   Obesity: 30.0-34.9
*   Underweight: < 18.5



```python
# Menghitung BMI
df_g['BMI'] = df_g['Berat Badan (Kg)'] / ((df_g['Tinggi Badan (Cm)'] / 100) ** 2)

# Membuat kategori BMI
def kategori_bmi(bmi):
    if bmi < 18.5:
        return 'Kurus'
    elif 18.5 <= bmi < 24.9:
        return 'Normal'
    elif 25.0 <= bmi < 29.9:
        return 'Gemuk'
    else:
        return 'Obesitas'

# Menambahkan kolom 'Status BMI' berdasarkan kategori BMI
df_g['Status BMI'] = df_g['BMI'].apply(kategori_bmi)

# Menampilkan dataframe dengan kolom 'Nama', 'BMI', dan 'Status BMI'
pd.DataFrame(df_g)
```

Penjelasan: code di atas bertujuan untuk mengetahui status BMI dari setiap peserta dan menyimpan hasilnya dalam kolom yang baru. untuk menjalankan program di atas, digunakan function untuk mengkategorikan nilai BMI nya (menggunakan if, elif, else untuk memenuhi kondisi yang diinginkan).

#### 9.Cari nilai-nilai berikut:

a. Kelompokkan tinggi badan dan berat badan serta cari nilai terendah dan tertingginya

```python
# Menggunakan groupby untuk mengelompokkan berdasarkan tinggi badan dan berat badan
grouped_data = df_g.groupby(['Tinggi Badan (Cm)', 'Berat Badan (Kg)']).size().reset_index(name='Jumlah')

# Mencari nilai terendah dan tertinggi dalam setiap kelompok
nilai_terendah = grouped_data.groupby(['Tinggi Badan (Cm)', 'Berat Badan (Kg)']).size().reset_index(name='Jumlah').min()
nilai_tertinggi = grouped_data.groupby(['Tinggi Badan (Cm)', 'Berat Badan (Kg)']).size().reset_index(name='Jumlah').max()

# Menggabungkan hasilnya dalam satu DataFrame
result = pd.DataFrame({'Nilai Terendah': nilai_terendah, 'Nilai Tertinggi': nilai_tertinggi})

# Menampilkan hasil
display(result)
```
